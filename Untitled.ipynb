```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.cos(x)

plt.figure(figsize=(8,4))
plt.plot(x,y,label = 'cos(x)', color = 'b')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('cos wave')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# lists can be modified while tuples cannot be modified
# Example of modifying a list
my_list = [1, 2, 3, 4]
print("Original List:", my_list)

# Modifying the list
my_list[2] = 99  # Changing the third element
my_list.append(5)  # Adding a new element
print("Modified List:", my_list)

# Example of a tuple (immutable)
my_tuple = (1, 2, 3, 4)
print("\nOriginal Tuple:", my_tuple)

# Attempting to modify the tuple
try:
    my_tuple[2] = 99  # This will raise a TypeError
except TypeError as e:
    print("Error:", e)

# Adding elements to a tuple (not allowed)
try:
    my_tuple += (5,)  # This creates a new tuple, not modifies the original
    print("Modified Tuple:", my_tuple)
except TypeError as e:
    print("Error:", e)
```
